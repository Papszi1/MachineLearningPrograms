```python
import numpy as np
import polars as pl
```

```python
lf = pl.scan_csv("abalone_data.csv")
```

```python
print(lf.collect_schema().names())
```

```python
lf_age = (
    lf.with_columns(
        (pl.col("Rings") + 1.5).alias("Age")
    )
)
print(lf_age.head().collect())
```

```python
rings_stats = lf_age.select(
    (
        (pl.col("Rings").mean()).alias("mean"),
        (pl.col("Rings").std()).alias("std"),
        (pl.col("Rings").var().alias("variance"))
    )
).collect()
age_stats = lf_age.select(
    (
        (pl.col("Age").mean()).alias("mean"),
        (pl.col("Age").std()).alias("std"),
        (pl.col("Age").var()).alias("variance")
    )
).collect()

print(rings_stats, age_stats)
```
