```python
import os
import numpy as np
```

```python
caffeines = np.random.uniform(0, 8, size=(100))
productivity = 10 + 1.5 * caffeines + np.random.normal(loc=1, scale=1, size=(100))
```

### calculate 
mean, variance, standard deviation, covariance, correlation

```python
mean_caffeines = np.sum(caffeines) / len(caffeines)
mean_productivity = np.sum(productivity) / len(productivity)
print(mean_caffeines, mean_productivity)
```

```python
variance_caffeines = 0
for value in caffeines:
    variance_caffeines += (value - mean_caffeines) ** 2
variance_caffeines /= len(caffeines)

variance_productivity = 0
for value in productivity:
    variance_productivity += (value - mean_productivity) ** 2
variance_productivity /= len(productivity)
print(variance_caffeines, variance_productivity)
```

```python
std_caffeines = variance_caffeines ** 0.5
std_productivity = variance_productivity ** 0.5
print(std_caffeines, std_productivity)
```

```python
cov_caff_prod = 0
for i in range(len(caffeines)):
    cov_caff_prod += (caffeines[i] - mean_caffeines)*(productivity[i] - mean_productivity)
    
cov_caff_prod /= len(caffeines)
print(cov_caff_prod)
```

```python
corr_caff_prod = cov_caff_prod / (std_caffeines * std_productivity)
print(corr_caff_prod)
```
